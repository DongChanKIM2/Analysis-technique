## 라이브러리 생성

```python
# 모듈 호출
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import collections
import itertools
from wordcloud import WordCloud
```

```python
df = pd.read_csv('연관어_2019.csv',encoding='CP949')
df.head(1)
```

## itertools를 활용해 검색어수 Counting

```python
kw_dict = collections.Counter(itertools.chain.from_iterable(k.split(',') for k in df['키워드']))
kw_dict
```

```python
rl_dict = collections.Counter(itertools.chain.from_iterable(k.split(',') for k in df['연관어']))
rl_dict
```

```python
kw_df = pd.DataFrame(list(kw_dict.items()),columns = ['word','count'])
kw_df.sort_values(by=['count'], ascending=False).head(20)
```
